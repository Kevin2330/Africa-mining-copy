```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Load the GeoJSON file for villages
geojson_path = '/Users/kd6801/Desktop/Mining-Project/Congo-Data/Congo-villages.geojson'
villages = gpd.read_file(geojson_path)

# Load the CSV file for existing mines
csv_path = '/Users/kd6801/Desktop/congo_existing_mines.csv'

try:
    mines_df = pd.read_csv(csv_path, encoding='utf-8')
except UnicodeDecodeError:
    mines_df = pd.read_csv(csv_path, encoding='latin1')

# Convert the mines DataFrame to a GeoDataFrame
mines_gdf = gpd.GeoDataFrame(
    mines_df, 
    geometry=gpd.points_from_xy(mines_df.longitude, mines_df.latitude),
    crs="EPSG:4326"
)

# Plot the data
fig, ax = plt.subplots(figsize=(10, 10))
villages.plot(ax=ax, color='blue', markersize=1, label='Villages')
mines_gdf.plot(ax=ax, color='red', markersize=5, label='Mines', marker='o')

# Customize the plot
ax.set_title('Village and Mine Distribution')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
